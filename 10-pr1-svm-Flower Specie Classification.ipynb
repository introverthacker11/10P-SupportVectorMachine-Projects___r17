###
## Creating DataFrame:
###

```python
import pandas as pd

df = pd.DataFrame(iris.data, columns = iris.feature_names)
df['target'] = iris.target
df['flower_name'] = df.target.apply(lambda x: iris.target_names[x])

df
```

###
## Checking Whether to Use Linear or Non-Linear SVM using Pair Plot:
###

```python
import seaborn as sns

colors = {'setosa' : 'green', 'versicolor' : 'blue', 'virginica' : 'red'}
sns.pairplot(df, hue = 'flower_name', palette = colors)
```

**Therefore, Based on the above pair plot, a non-linear SVM would be a more appropriate choice for this dataset.**

**Here's why:**

- Overlapping Classes: The scatter plots show significant overlap between the different flower species (setosa, versicolor, virginica), indicating that a linear decision boundary would not be able to effectively separate them.
* Complex Relationships: The relationships between features (sepal length, sepal width, petal length, petal width) appear to be non-linear, as evidenced by the scatter plots.

###
## Seperating Values:
###


```python
x = df.drop(['target', 'flower_name'], axis = 1)
x
```

```python
y = df['target']
y
```

###
## train_test_split:
###

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

###
## Shapes:
###

```python
x.shape, y.shape, x_train.shape , x_test.shape , y_train.shape , y_test.shape
```

###
## Importing Support Vector Machine:
###

```python
from sklearn.svm import SVC

model = SVC()
```

###
## Training Algorithm:
###

```python
model.fit(x_train, y_train)
```

###
## Predicting using Unseen Dataset, i.e: test dataset (x_test):
###

```python
model.predict(x_test)
```

###
## Model Accuracy:
###

```python
print(f'Model Acuuracy Rate: {model.score(x_test, y_test) * 100:.3f} %')
```

###
## Orignal vs Predicted Values:
###

```python
pd.DataFrame({'Orignal Values' : y_test, 'Predicted Values' : model.predict(x_test)})
```
